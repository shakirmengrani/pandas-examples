```python
import requests, json, matplotlib.pyplot as plt
import pandas as pd, numpy as np
```

```python
def show(df):
    market_sum = np.sum(df["price"])
    market_mean = np.true_divide(market_sum, len(df["market"]))
    print("\nMarket Sum: {} and it's Mean: {}".format(market_sum, market_mean))
    fig, ax = plt.subplots()
    fig.set_size_inches(20,8)
    ax.plot(df["market"], df["price"])
    ax.set_title("Market Rate", fontsize=16)
    ax.set_xlabel("Market", fontsize=16)
    ax.set_ylabel("Pice", fontsize=16)
    plt.show()
    print("Before starting trade we must focus on Market Mean: {}".format(market_mean))
```

```python
def fetch():
    r = requests.get("https://api.cryptonator.com/api/full/btc-usd")
    markets = r.json()["ticker"]["markets"]
    df = pd.read_json( json.dumps(markets) )
    print(df)
    show(df)
```

```python
fetch()
```
